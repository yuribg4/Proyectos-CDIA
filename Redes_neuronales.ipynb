**Perceptron**

Es un concepto que nos permiten tomar descisiones , es como el arbol de decicsiones en este caso si la respuesta es positiva es 1 y si es negativa 0

**Umbral**

los datos anteriores se procesan aqui , si el valor es mayo que la umbral el resultado sera positio y si no sera negativo

**Pesos(Weights)**

A cada decision se le asigna un valor dependiendo de su importancia antes de entrar al perceptron las decisiones se muliplican por su importancia

**Perceptron Multicapa(Red Neuronal)**

 Son varios perceptrones en varias capas cada una conectanco con otro percepton para poder llegar a una sola decision

```python
import pandas as pd

datos = pd.read_csv("/content/celsius - copia.csv")
```

```python
datos.head()
```

```python
datos['celsius'].values
```

```python
X = datos['celsius'].values
y = datos['fahrenheit'].values
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

modelo.compile(
    optimizer= tf.keras.optimizers.Adam(.1),
    loss="mean_squared_error"
)
```

```python
entrenamiento = modelo.fit(X,y, epochs=1000)
```

```python
import matplotlib.pyplot as plt

plt.xlabel("# Epoca")
plt.ylabel("Magnitud de la perdida")
plt.plot(entrenamiento.history["loss"])
```

```python
import numpy as np

c = 1230
prediccion = modelo.predict(np.array([[c]]))
print (f"Prediccion : {c} celcius son {prediccion[0]} farenheit")
```

```python
print(modelo.layers[0].get_weights())
```
