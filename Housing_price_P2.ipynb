```python
#separar los caracteristicas de la etiqueta
X = datos_na.drop(["median_house_value"],axis=1)
y= datos_na["median_house_value"]
```

```python
#separa en 2 parte: entrenamiento y prueba
from sklearn.model_selection import train_test_split
X_ent, X_pru, y_ent, y_pru = train_test_split(X, y, test_size=0.2)
```

```python
y_ent.shape
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
```

```python
# Convert all categorical columns to dummy variables
X_ent = pd.get_dummies(X_ent)
X_pru = pd.get_dummies(X_pru)
```

```python
modelo.fit(X_ent, y_ent)
```

```python
predicciones = modelo.predict(X_pru)
```

```python
comparativa = {"Prediccion" : predicciones, "Valor Real" : y_pru}
pd.DataFrame(predicciones)
```

```python
#Overffiting sobreajuste
print(modelo.score(X_ent,y_ent))
print(modelo.score(X_pru,y_pru))
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_pru,predicciones)
```

```python
mse
```

```python
rmse=np.sqrt(mse)
```
