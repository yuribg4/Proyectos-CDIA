```python
#Librerias
import requests
from bs4 import BeautifulSoup
import re
```

```python
# 1. Obtenemos el HTML de la página (usaremos una de prueba común)
url = "https://elfondoenlinea.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```

```python
# Buscamos todos los contenedores de libros (ejemplo de clase genérica)
libros_title = soup.find_all('p', class_='titulo-mini-slider')

for libro in libros_title:
 # Usamos find para el primer título que encuentre dentro de ese libro
    titulo = libro.text.strip()
    print(f"Libro encontrado: {titulo}")

libros_autor = soup.find_all('p', class_='autor-mini-slider')
for libro in libros_autor:
 # Usamos find para el primer título que encuentre dentro de ese libro
    Autor = libro.text.strip()
    print(f"Autor Encontrado: {Autor}")
libros_precio = soup.find_all('ul', class_='nav fce-precios')
for libro in libros_precio:
 # Usamos find para el primer título que encuentre dentro de ese libro
    precio = libro.text.strip()
    print(f"Precio Encontrado: {precio}")
```

```python
import pandas as pd
```

```python
# Extraer títulos
libros_title = soup.find_all('p', class_='titulo-mini-slider')
titulos = [libro.text.strip() for libro in libros_title]

# Extraer autores
libros_autor = soup.find_all('p', class_='autor-mini-slider')
autores = [libro.text.strip() for libro in libros_autor]

# Extraer precios
libros_precio = soup.find_all('ul', class_='nav fce-precios')
precios = [libro.text.strip() for libro in libros_precio]

# Crear arreglo de diccionarios
libros = []

for i in range(len(titulos)):
    libro = {
        "Titulo": titulos[i],
        "Autor": autores[i] if i < len(autores) else None,
        "Precio": precios[i] if i < len(precios) else None
    }
    libros.append(libro)

# Convertir a DataFrame
df = pd.DataFrame(libros)

# Mostrar resultado
df
```
